```r
#build a grid of parameters to optimize (each row will be used to genereate a candidate model for evaluation)
grid <- expand.grid(.model = "tree", .trials = c(1,5,10,15,20,25,30,35), .winnow = FALSE)
grid
```

```r
#Random Forests
#Training random forests
install.packages("randomForest", repos = "https://cran.r-project.org")
library(randomForest)
```

```r
set.seed(300)
rf <- randomForest(default ~ ., data = credit)
rf
```

```r
#Evaluating random forest performance (auto-tuned random forest vs auto-tuned boosted C5.0)
ctrl <- trainControl(method = "repeatedcv", number = 10, repeats = 10)
```

```r
#Random forest
grid_rf <- expand.grid(.mtry = c(2, 4, 8 ,16))
grid_rf
set.seed(300)
m_rf <- train(default ~ ., data = credit, method = "rf", metric = "Kappa", trControl = ctrl, tuneGrid = grid_rf)
```

```r
#C5.0
grid_c50 <- expand.grid(.model = "tree", .trials = c(10, 20, 30, 40), .winnow = "FALSE")
set.seed(300)
m_c50 <- train(default ~ ., data = credit, method = "C5.0", metric = "Kappa", trControl = ctrl, tuneGrid = grid_c50)
```
